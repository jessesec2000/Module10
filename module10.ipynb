```python
import matplotlib
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *

Cities = ["Lome", "Yaounde", "Djamenna", "Ouaga", "Dakar", "Lesotho", "Accra", "Nairobi"]
MMR = [14, 21, 21, 19, 20, 16, 17, 18] #Maternal Mortality Rate rounded up to the nearst whole number
df = pd.DataFrame(data=zip(Cities, MMR), columns=['Cities', 'MMR'])

(ggplot(df)
 + aes (x='Cities', weight='MMR')
 + geom_bar (size=15)
 + ylab('Mother Mortality Rate')
 + ggtitle ('Mother Mortality Rate in African Cities')
)
```
